```python
f = open('transcript.txt','r')
text = f.read().replace("?",".")
script = text.split(".")
script = [e.replace("\n","") + "." for e in script]
script = script[:-1]
script
```

```python
#speakers = ['tucker-carlson','david-attenborough','ben-shapiro']
#speakers = ['ben-shapiro']
speakers = ['david-attenborough']
results = []
```

```python
for i in range(len(script)):
    speaker = speakers[i%len(speakers)]
    result = query_utils.create_audio(script[i],speaker,"AudioResults\\" +  str(i) + ".wav")
    results.append(result)
    time.sleep(2)
```

```python
import wave

x = ["AudioResults\\" + str(i) + ".wav" for i in range(len(script))]
infiles = []
for e in x:
    infiles.append(e)
    infiles.append("AudioResults\\wait.wav")
outfile = "a.wav"

data= []
for infile in infiles:
    w = wave.open(infile, 'rb')
    data.append( [w.getparams(), w.readframes(w.getnframes())] )
    w.close()
    
output = wave.open(outfile, 'wb')

output.setparams(data[0][0])
for i in range(len(data)):
    output.writeframes(data[i][1])

output.close()
```
