```python
import numpy as np
from PIL import Image, ImageSequence

img = Image.open('basefile.gif')
frames = np.array([np.array(frame.copy().convert('RGB').getdata(),dtype=np.uint8).reshape(frame.size[1],frame.size[0],3) for frame in ImageSequence.Iterator(img)])
```

```python
talk_durations = [3,2]
final_frames = []
n_frames = len(frames)
```

```python
cur = 0
for e in talk_durations:
    for i in range(e * 10):
        final_frames.append(frames[i%n_frames])
    for i in range(5):
        final_frames.append(frames[0])
```

```python
final_frames = np.array(final_frames)
```

```python
write_gif(final_frames,'test.gif',fps=10)
```

```python
import cv2
import numpy as np
from cv2 import VideoWriter, VideoWriter_fourcc
```

```python
FPS = 10
width, height = 500, 359
```

```python
fourcc = VideoWriter_fourcc(*'MP42')
video = VideoWriter('./noise.avi', fourcc, float(FPS), (width, height))
```

```python
for frame in final_frames:
    video.write(frame)
```

```python
video.release()
```
